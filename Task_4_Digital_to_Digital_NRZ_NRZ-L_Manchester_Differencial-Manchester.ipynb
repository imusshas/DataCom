# Digital-to-Digital Conversion Techniques

This notebook simulates digital-to-digital conversion techniques such as Non-Return-to-Zero (NRZ),
Non-Return-to-Zero-Level (NRZ-L), Manchester, and Differential Manchester encoding.

## 1. Generating a Digital Signal

Create a random binary signal.


```python
import numpy as np
import matplotlib.pyplot as plt


num_bits = 8
bit_rate = 1  # bits per second
time = np.linspace(0, num_bits, num_bits * 100)

# Generate a random binary signal
digital_signal = np.random.choice([0, 1], size=num_bits)

# Visualize the digital signal
plt.figure(figsize=(10, 4))
plt.step(range(num_bits), digital_signal, where='post', label='Digital Signal')
plt.xlabel('Bit Position')
plt.ylabel('Amplitude')
plt.title('Generated Digital Signal')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2. Digital-to-Digital Conversion Techniques

### 2.1. Non-Return-to-Zero (NRZ)

In NRZ encoding, a '1' is represented by a high level and a '0' by a low level.


```python
nrz_signal = np.repeat(digital_signal, 100)

# Visualize the NRZ encoded signal
plt.figure(figsize=(10, 4))
plt.plot(time, nrz_signal, label='NRZ Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Non-Return-to-Zero (NRZ) Signal')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2.2. Non-Return-to-Zero-Level (NRZ-L)

In NRZ-L encoding, the level does not return to zero between bits, and '0' and '1' are represented by different voltage
levels.


```python
nrzl_signal = np.repeat(digital_signal, 100)

# Visualize the NRZ-L encoded signal
plt.figure(figsize=(10, 4))
plt.plot(time, nrzl_signal, label='NRZ-L Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Non-Return-to-Zero-Level (NRZ-L) Signal')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2.3. Manchester Encoding

In Manchester encoding, each bit has a transition in the middle: '0' is high-to-low, and '1' is low-to-high.


```python

manchester_signal = np.zeros(num_bits * 100)
for i in range(num_bits):
    manchester_signal[i*100:(i+1)*100//2] = digital_signal[i]
    manchester_signal[(i+1)*100//2:(i+1)*100] = 1 - digital_signal[i]

# Visualize the Manchester encoded signal
plt.figure(figsize=(10, 4))
plt.plot(time, manchester_signal, label='Manchester Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Manchester Encoding')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 2.4. Differential Manchester Encoding

In Differential Manchester encoding, each bit has a transition in the middle, and a transition at the beginning of a bit
period represents '0', while no transition at the beginning represents '1'.


```python
diff_manchester_signal = np.zeros(num_bits * 100)
last_bit = 1
for i in range(num_bits):
    if digital_signal[i] == 0:
        diff_manchester_signal[i*100:(i+1)*100//2] = last_bit
        diff_manchester_signal[(i+1)*100//2:(i+1)*100] = 1 - last_bit
    else:
        last_bit = 1 - last_bit
        diff_manchester_signal[i*100:(i+1)*100//2] = last_bit
        diff_manchester_signal[(i+1)*100//2:(i+1)*100] = 1 - last_bit
    last_bit = diff_manchester_signal[(i+1)*100-1]

# Visualize the Differential Manchester encoded signal
plt.figure(figsize=(10, 4))
plt.plot(time, diff_manchester_signal, label='Differential Manchester Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Differential Manchester Encoding')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```
