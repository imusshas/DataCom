# Digital-to-Analog Conversion Techniques

This simulates digital to analog conversion techniques such as Amplitude Shift Keying (ASK), Frequency Shift Keying
(FSK), and Phase Shift Keying (PSK).

## 1. Digital-to-Analog Conversion Techniques

Generate a random digital signal to modulate the analog signal using ASK, FSK, and PSK techniques.

Generate a Digital Signal


```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters for digital signal
num_bits = 8
bit_duration = 0.1  # seconds
digital_signal = np.random.choice([0, 1], num_bits)
sampling_rate = 1000  # Hz
amplitude = 1  # V

# Create a time array for the digital signal
time_digital = np.linspace(0, num_bits * bit_duration, num_bits * sampling_rate // 10)

# Upsample the digital signal for modulation
digital_signal_upsampled = np.repeat(digital_signal, sampling_rate // 10)

# Visualize the digital signal
plt.figure(figsize=(10, 4))
plt.step(time_digital, digital_signal_upsampled, where='post', label='Digital Signal')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')
plt.title('Generated Digital Signal')
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 1.1. Amplitude Shift Keying (ASK)

In ASK, the amplitude of the carrier signal is varied according to the digital signal.


```python

carrier_frequency = 10  # Hz
carrier_wave = amplitude * np.sin(2 * np.pi * carrier_frequency * time_digital)

# Modulate the carrier wave using ASK
ask_signal = (digital_signal_upsampled * 2 - 1) * carrier_wave

# Visualize the ASK signal
plt.figure(figsize=(10, 4))
plt.plot(time_digital, ask_signal, label='ASK Signal')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')
plt.title('Amplitude Shift Keying (ASK) Signal')
plt.grid(True)
plt.legend()
plt.show()
```

## 1.2. Frequency Shift Keying (FSK)

In FSK, the frequency of the carrier signal is varied according to the digital signal.


```python

frequency1 = 5  # Hz for bit 0
frequency2 = 15  # Hz for bit 1

# Generate FSK signal
fsk_signal = np.zeros_like(time_digital)
for i in range(num_bits):
    if digital_signal[i] == 0:
        fsk_signal[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10] = amplitude * np.sin(2 * np.pi * frequency1 * time_digital[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10])
    else:
        fsk_signal[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10] = amplitude * np.sin(2 * np.pi * frequency2 * time_digital[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10])

# Visualize the FSK signal
plt.figure(figsize=(10, 4))
plt.plot(time_digital, fsk_signal, label='FSK Signal')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')
plt.title('Frequency Shift Keying (FSK) Signal')
plt.grid(True)
plt.legend()
plt.show()
```

## 1.3. Phase Shift Keying (PSK)

In PSK, the phase of the carrier signal is varied according to the digital signal.


```python
# Parameters for PSK
psk_signal = np.zeros_like(time_digital)

# Generate PSK signal
for i in range(num_bits):
    if digital_signal[i] == 0:
        psk_signal[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10] = amplitude * np.sin(2 * np.pi * carrier_frequency * time_digital[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10])
    else:
        psk_signal[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10] = amplitude * np.sin(2 * np.pi * carrier_frequency * time_digital[i * sampling_rate // 10 : (i + 1) * sampling_rate // 10] + np.pi)

# Visualize the PSK signal
plt.figure(figsize=(10, 4))
plt.plot(time_digital, psk_signal, label='PSK Signal')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')
plt.title('Phase Shift Keying (PSK) Signal')
plt.grid(True)
plt.legend()
plt.show()
```
